```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_2014 = pd.read_csv("../data/clean/2014.csv")
df_2020 = pd.read_csv("../data/clean/2020.csv")
```

```python
df_inc_chg = pd.DataFrame({'INC_CHG': df_2020['MEDINC'] - df_2014['MEDINC'],
                          'POP_CHG': df_2020['TOT_POP']- df_2014['TOT_POP']}).sort_values(by='INC_CHG')
```

```python
from sklearn.linear_model import LinearRegression

x = df_inc_chg['INC_CHG'].to_numpy().reshape(-1,1)
y = df_inc_chg['POP_CHG'].to_numpy()#.reshape(1,-1)

pop_model = LinearRegression()
pop_model.fit(x,y)
y_predict = pop_model.predict(x)
```

```python
plt.scatter(df_inc_chg['INC_CHG'], df_inc_chg['POP_CHG'], s=12)
plt.plot(x, y_predict)
plt.xlabel("Median Income Change ($)")
plt.ylabel("Population change (# people)")
plt.title("Population change as function of \nincome change, by neighborhood")
```

### Explanation

 Here, I show how the population has changed in each neighborhood plotted against the change in median income from 2014 to 2020 and show a simple linear regression to fit the data. This indcates, not surprisingly, a positive correlation--increased income is correlated with increased population growth
